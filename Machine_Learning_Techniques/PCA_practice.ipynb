```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
rng = np.random.RandomState(0)
n_samples = 500
cov = [[3, 3],
       [3, 4]]
X = rng.multivariate_normal(mean=[0, 0], cov=cov, size=n_samples)
pca = PCA(n_components=2).fit(X)
```

```python
pca_features = pca.transform(X)
plt.scatter(pca_features[:, 0], pca_features[:, 1], alpha=.3, label='pca_features')
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=.3, label='samples')
```

```python
pca.explained_variance_[0]
```

```python
pca.components_[0]*pca.explained_variance_[0]
```

```python
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
print(np.var(pca_features[:, 0]))
print(np.var(pca_features[:, 1]) )
```

```python

for i, (comp, var) in enumerate(zip(pca.components_, pca.explained_variance_)):
    comp = comp * var  # scale component by its variance explanation power
    plt.plot([0, comp[0]], [0, comp[1]], label=f"Component {i}", linewidth=5,
             color=f"C{i + 2}")
plt.gca().set(aspect='equal',
              title="2-dimensional dataset with principal components",
              xlabel='first feature', ylabel='second feature')
plt.legend()
plt.show()
```
